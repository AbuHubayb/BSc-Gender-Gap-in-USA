## This project intends to visualize the gap between the male and female BSc graduates of colleges across the years 1970 to 2011.

**The data was taken from The Department of Education Statistics data set released annually on http://nces.ed.gov/programs/digest/2013menu_tables.asp ***

The outcome should present answers to the following questions;

        1. What was the level of involvement/awareness of women as regards the attainment of a Bachelor's degree?
        2. Has there been any changes over the years?
        3. What is the trend across various disciples?
        4. What's the future like for individual courses? Are there chances that women would get more involved than men?

### Read in the file

```python
# import the necessary modules and libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

The colors for the line plots should be called now

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

Presenting the courses in categories would be an easier way to visualize them

```python
stem_cats =     ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
                 'Social Sciences and History']
other_cats =    ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

### Making-out the plot

```python
fig = plt.figure(figsize=(18, 18))


# Plotting for the first category, stem_cats

for i in range(0,6): 
    ax = fig.add_subplot(6,3,(3 * i) + 1, xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
#     plot for men's degrees would be 100-women's
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c=cb_orange, label='Men', linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[i])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if i == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
        

# Ploting for the second category, library arts category

for i in range(0,5): 
    ax = fig.add_subplot(6,3,(3 * i) + 2, xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[i]], c=cb_orange, label='Men', linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[i])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if i == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
        
                
# Ploting for the last category, other_cat

for i in range(0,6): 
    ax = fig.add_subplot(6,3,(3 * i) + 3, xticklabels=[], yticklabels=[], xticks=[], yticks=[])
    ax.plot(women_degrees['Year'], women_degrees[other_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c=cb_orange, label='Men', linewidth=3)
       
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[i])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if i == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

plt.show()
# fig.savefig('Women-Men Degrees.png', format='png')
```
